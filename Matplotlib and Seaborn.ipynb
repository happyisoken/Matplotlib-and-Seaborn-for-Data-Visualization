### Pie Charts

A Pie Chart(or a circle chart) is a type of graph that displays data in a circular graph. It is a circular statistical graphic which is divided into slices to illustrate numerical proportions. The pieces of the graph are proportional to the fraction of the whole in each category. In other words, each slice of the pie is relative to the size of that category in the group as a whole. The entire pie represents 100 percent of a whole, while the pie slices represent portions of the whole.

```python
# Pie Chart Number 2
# the strip() method removes any spaces or specified characters at the start and end of a string.
# the int function converts the result to integer.
fifa.Weight = [int(x.strip('lbs')) if type (x)== str else x for x in fifa.Weight]

fifa.Weight[0]              
```

```python
# To get the weight distribution of FIFA Players 
# The strip() method removes any spaces or specified characters at the start and end of a string

fifa.Weight = [int(x.strip('lbs')) if type (x)== str else x for x in fifa.Weight]

# plt.style.use('ggplot') - to give a default style to your chart

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[(fifa.Weight >= 200)].count()[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125 - 150', '150 - 175', '175 - 200', 'Over 200']
colors = ['#0c0d0c', '#429bf5', '#d142f5', '#b3f542', '#d9092f']
explode = (.4, .2, 0, 0, .4)

plt.figure(figsize = (8, 6))

plt.title ('Weight Distribution of FIFA Players (in lbs)')          

plt.pie(weights, labels = labels, colors = colors, autopct = '%.2f %%', pctdistance = 0.8, explode = explode) # %.2f means to 2 decimal places while pctdistance is distance away from the centre

plt.show()
```

### Box Plot

A box and whisker plotalso called a box plotdisplays the five-number summary of a set of data. The five-number summary is the minimum, first quartile, median, third quartile, and maximum. In a box plot, we draw a box from the first quartile to the third quartile. A vertical line goes through the box at the median.

```python
# Using box Plot to compare team Barcelona, Real Madrid and New England Revolution
plt.style.use('default')

plt.figure(figsize = (5, 8))

barcelona = fifa.loc[fifa.Club =='FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ['FC Barcelona', 'Real Madrid', 'New England Revolution']

boxes = plt.boxplot([barcelona, madrid, revs], labels = labels, patch_artist = True, medianprops = {'linewidth':2})

for box in boxes['boxes']:
    # set edge color
    box.set(color = '#2432f0', linewidth = 2)
    
    # change Fill color
    box.set(facecolor = '#d8d9ed')

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

plt.show()
```

### PairGrid

PairGrid allows us to draw a grid of subplots using the same plot type to visualize data. It uses different pair of variable for each subplot. It forms a matrix of sub-plots. It is also sometimes called as scatterplot matrix. 

```python
# Load the iris dataset
iris = sns.load_dataset('iris')

# Then create an empty grid system based on the iris data
iris_g = sns.PairGrid(iris, hue = 'species')

# Then put a scatter plot accross the upper, lower and the diagonal
iris_g.map(plt.scatter)
```

```python
# Creating a histograms in the scatter plots
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris, hue = 'species')
iris_g.map(plt.scatter)
iris_g.map_diag(plt.hist)
```

```python
# Removing the scatter plots where the histograms are
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris, hue = 'species')
iris_g.map_diag(plt.hist)
iris_g.map_offdiag(plt.scatter)
```

```python
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris, hue = 'species')
iris_g.map_diag(plt.hist)
iris_g.map_upper(plt.scatter)
iris_g.map_lower(sns.kdeplot)
```

```python
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris, hue = 'species',
                     x_vars = ['sepal_length', 'sepal_width'],
                     y_vars = ['petal_length', 'petal_width'])
iris_g.map(plt.scatter)
iris_g.add_legend()
```

### Facet Grid


FacetGrid object takes a dataframe as input and the names of the variables that will form the row, column, or hue dimensions of the grid. The variables should be categorical and the data at each level of the variable will be used for a facet along that axis.

```python
# Creating a grid for smokers and non smokers
tips_fg = sns.FacetGrid(tips_df, col = 'time', row = 'smoker')
tips_fg.map(plt.hist, 'total_bill', bins = 8)
```

```python
tips_fg = sns.FacetGrid(tips_df, col = 'time', row = 'smoker')
tips_fg.map(plt.scatter, 'total_bill', 'tip')
```

```python
# Assigning variables to different colors
tips_fg = sns.FacetGrid(tips_df, col = 'time', hue = 'smoker', height = 4, aspect = 1.3)
tips_fg.map(plt.scatter, 'total_bill', 'tip')

tips_fg.add_legend()
```

```python
# Making dinner appear before lunch
tips_fg = sns.FacetGrid(tips_df, col = 'time', hue = 'smoker', height = 4, aspect = 1.3, col_order = ['Dinner','Lunch'], palette = 'Set1')
tips_fg.map(plt.scatter, 'total_bill', 'tip', edgecolor = 'w')
```

```python
kws = dict(s = 200, linewidth = 5, edgecolor = 'w')
tips_fg = sns.FacetGrid(tips_df, col = 'sex', hue = 'smoker', height = 4, aspect = 1.3,
                         hue_order = ['Yes', 'No'],
                         hue_kws = dict(marker = ['^', 'v']))
tips_fg.map(plt.scatter, 'total_bill', 'tip', **kws)
```
